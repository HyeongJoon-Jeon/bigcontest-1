```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
korean = pd.read_csv("C:\\Users\\dayfl\\Desktop\\2021 빅콘테스트_데이터분석분야_퓨처스리그_ECO제주_데이터_210806_update\\01_제공데이터\\02-1_내국인유동인구_KOREAN.csv", encoding = "cp949")
```

```python
korean
```

```python
date = []
for i in range(len(korean)):
    date.append(korean["base_date"][i][:7])
```

```python
korean_date = pd.Series(date, name = "base_date")
```

```python
quarter = []
```

```python
quarter = []
for i in range(len(korean)):
    if 1 <= korean["time"][i] <= 6:
        quarter.append(1)
    elif 7 <= korean["time"][i] <= 12:
        quarter.append(2)
    elif 13 <= korean["time"][i] <= 18:
        quarter.append(3)
    else:
        quarter.append(4)
```

```python
korean_time = pd.Series(quarter, name = "time")
```

```python
korean_time
```

```python
generation = []
for i in range(len(korean)):
    if 0 <= korean["age"][i] <= 10:
        generation.append(1)
    elif 20 <= korean["age"][i] <= 30:
        generation.append(2)
    elif 40 <= korean["age"][i] <= 50:
        generation.append(3)
    elif 60 <= korean["age"][i] <= 80:
        generation.append(4)
```

```python
korean_age = pd.Series(generation, name = "age")
```

```python
korean_age
```

```python
korean1= korean[["emd_nm", "sex"]]
```

```python
korean1
```

```python
korean2 = korean[["resd_pop_cnt", "work_pop_cnt", "visit_pop_cnt"]]
```

```python
korean2
```

```python
korean3 = pd.concat([korean_date, korean_time, korean1, korean_age, korean2], axis = 1)
```

```python
korean3
```

```python
korean4 = korean3.groupby(['base_date', 'time', 'emd_nm', 'sex', 'age']).sum()
```

```python
korean4
```

```python
korean5 = korean3.groupby(['base_date', 'time', 'emd_nm', 'sex', 'age']).sum().reset_index()
```

```python
korean5
```

```python
korean5.to_csv('korean.csv')
```
