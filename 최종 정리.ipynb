```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
merge1 = pd.read_csv("merge1.csv", encoding='utf-8-sig')
merge2 = pd.read_csv("merge2.csv", encoding='utf-8-sig')
merge3 = pd.read_csv("merge3.csv", encoding='utf-8-sig')
```

```python
merge1
```

```python
merge1 = merge1.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
merge1
```

```python
merge2
```

```python
merge3
```

```python
merge3 = merge3.drop(['Unnamed: 0'], axis=1)
```

```python
merge3
```

```python
merge = pd.concat((merge1, merge2, merge3), axis = 0)
```

```python
merge
```

```python
merge['base_date'] = pd.to_datetime(merge['base_date'])
```

```python
merge['year'] = merge['base_date'].dt.year
```

```python
merge['month'] = merge['base_date'].dt.month
```

```python
merge = merge.groupby(['year','month', 'emd_nm'])['em_g'].sum().reset_index()
```

```python
merge.to_csv("제출용.csv", encoding='utf-8-sig', index=None)
```
